Perform the T test to check the caim of the production is going to happen for this cost with significance level alpha 0.05,sample mean=500 and dataSet has sample of 40 entries

Hypthesis(2 tailed test)

NULL hypthesis
Average cost is equal to 500 crore


alternative hypothesis
Average cost is not equal to 500 crore

```python
from scipy import stats
```

```python
stats.ttest_1samp(df["production_cost"],500)
```

Feature Scaling

1.Absolute MAximum Scaling

```python
import pandas as pd
```

```python


f = pd.DataFrame({"LoArea": [8450, 9600, 11250, 9550, 14620], "MSSubclass": [60, 20, 60, 70, 60]})
```

```python
f
```

```python
# finding the absolute value
import numpy as np
max_val=np.max(np.abs(f))
```

```python
max_val
```

```python
print((f-max_val)/max_val)
```

2.Mean Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
sacled_data=scaler.fit_transform(f)
scaled_df=pd.DataFrame(sacled_data,columns=f.columns)
```

```python
scaled_df
```

Normalisation

```python
from sklearn.preprocessing import Normalizer
scaler=Normalizer()
sacled_entry=scaler.fit_transform(f)
scaled_norm=pd.DataFrame(sacled_data,columns=f.columns)
```

```python
scaled_norm
```

Standardisation

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
sacled_data=scaler.fit_transform(f)
scaled_df=pd.DataFrame(sacled_data,columns=f.columns)
```

```python
scaled_df
```

Robust Scaling

```python
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
sacled_data=scaler.fit_transform(f)
scaled_df=pd.DataFrame(sacled_data,columns=f.columns)
```

```python
scaled_df
```

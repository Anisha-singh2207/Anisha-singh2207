```python
import pandas as pd
```

```python
df=pd.read_csv("/content/load_boston_final_lyst1729240745225.csv")
```

```python
df
```

```python
x=df.iloc[:,1:14]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
# sacling so that the different scales of the dataset entries become available
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score

c_v = cross_val_score(lr, x_train, y_train, scoring='neg_mean_squared_error', cv=10)
c_v_1=cross_val_score(lr,x_train,y_train,scoring='neg_mean_absolute_error',cv=10)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
lr.score(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

Ridge regularisation

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge_1=Ridge()
```

```python
parameters={"alpha":[1,2,3,4,5,6,7,10,30,40,50,70,45]}
ridge_cv=GridSearchCV(ridge_1,parameters,scoring="neg_mean_squared_error",cv=10)
ridge_cv.fit(x_train,y_train)
```

```python
ridge_cv.best_params_
```

```python
ridge_pred=ridge_cv.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(ridge_pred,y_test,)
```

```python
r2_score(y_test,y_pred)
```

Lasso Regression model:


```python
from sklearn.linear_model import Lasso
```

```python
lasso_l=Lasso()
```

```python
parameters={"alpha":[1,2,3,4,5,6,7,10,30,40,50,70,45]}
lasso_cv = GridSearchCV(lasso_l, parameters, scoring="neg_mean_squared_error", cv=5)
lasso_cv.fit(x_train,y_train)
```

```python
lasso_pred=lasso_cv.predict(x_test)
```

```python
r2_score(y_test,lasso_pred)
```
